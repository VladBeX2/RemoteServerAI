# Reparam Combined Corpus dataset 

```python
import pandas as pd
import glob

train_files = glob.glob("../datasets/Combined_Corpus/Train/*.csv")
test_files  = glob.glob("../datasets/Combined_Corpus/Test/*.csv")

dfs_train = []
dfs_test = []
for f in train_files:
    df_part = pd.read_csv(f, on_bad_lines='skip', engine='python')
    dfs_train.append(df_part)

df_train = pd.concat(dfs_train, ignore_index=True)

for f in test_files:
    df_part = pd.read_csv(f, on_bad_lines='skip', engine='python')
    dfs_test.append(df_part)

df_test = pd.concat(dfs_test, ignore_index=True)
df_all = pd.concat([df_train, df_test], ignore_index=True)
```

```python
df_all
```

```python
df_all["Label"] = pd.to_numeric(df_all["Label"], errors="coerce")

df_all = df_all[df_all["Label"].isin([0, 1])]

label_counts = df_all["Label"].value_counts()
print("Distribuția etichetelor (0/1) în Combined Corpus curățat:")
print(label_counts)
```
