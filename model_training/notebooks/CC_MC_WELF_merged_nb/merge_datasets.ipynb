```python
import pandas as pd
```

```python
df_mc = pd.read_csv('../../datasets/MC_Fake_cleaned.csv')
print(df_mc.sample(5))
```

```python
print(df_mc['label'].value_counts())
```

```python
df_cc = pd.read_csv('../../datasets/Combined_Corpus/All_cleaned.csv')
df_cc.sample(5)
```

```python
print(df_cc['label'].value_counts())
```

```python
df_wf = pd.read_csv('../../datasets/WELFake_cleaned.csv')
print(df_wf.sample(5))
print(df_wf['label'].value_counts())
```

```python
#we will be using CC labels (1 for real, 0 for fake)
df_mc['label'] = 1 - df_mc['label']
df_wf['label'] = 1 - df_wf['label']

#Make a column with the source of the dataset
df_mc['source'] = 'MC'
df_cc['source'] = 'CC'
df_wf['source'] = 'WF'
#Combine the datasets
df_combined = pd.concat([df_mc, df_cc, df_wf], ignore_index=True)

#keep only title, text, label, language, category, source
df_combined = df_combined[['title', 'text', 'label', 'language', 'category', 'source']]
```

```python
df_combined.sample(5)
```

```python
df_combined['label'].value_counts()
```

```python
#save the combined dataset
df_combined.to_csv('../../datasets/CC_MC_WELF_merged.csv', index=False)
```
