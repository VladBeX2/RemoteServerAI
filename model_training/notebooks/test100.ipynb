```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/teste100.tsv',sep='\t')
print(df.tail())
count_0 = df[df['label(1=real,0=fake)'] == 0].shape[0]
count_1 = df[df['label(1=real,0=fake)'] == 1].shape[0]
print(f'Count of label(1=real,0=fake) 0: {count_0}')
print(f'Count of label(1=real,0=fake) 1: {count_1}')
```

```python
df.isnull().sum()
which_null = df[df.isnull().any(axis=1)]
print(f'Rows with null values: {which_null.shape[0]}')
#whichones are null
print('Rows with null values:')
print(which_null.index.tolist())
```

```python
#verify duplicates
df.duplicated().sum()
#which are the duplicated rows?
duplicates = df[df.duplicated(keep=False)]
print(f'Number of duplicated rows: {duplicates.shape[0]}')
#show the duplicated rows
print(duplicates)
```
