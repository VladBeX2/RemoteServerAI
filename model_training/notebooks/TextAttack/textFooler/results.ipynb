```python
import pandas as pd
```

```python
r0 = pd.read_csv('test_textfooler_results_gpu_0.csv')
r1 = pd.read_csv('test_textfooler_results_gpu_1.csv')
r2 = pd.read_csv('test_textfooler_results_gpu_2.csv')
```

```python
# Merge the dataframes
print(r0.shape)
print(r1.shape)
print(r2.shape)
merged = pd.concat([r0, r1, r2], ignore_index=True)
print(merged.shape)
```

```python
merged.head()
```

```python
merged['result_type'].value_counts()
```

```python
#how many of the column original output is the same as ground truth
def get_same_as_gt(row):
    if row['original_output'] == row['ground_truth_output']:
        return 1
    else:
        return 0
    
#apply the function to the dataframe
merged['same_as_gt'] = merged.apply(get_same_as_gt, axis=1)
merged['same_as_gt'].value_counts()
```

```python
succes = merged[merged["result_type"] == "Successful"]

succes['ground_truth_output'].value_counts()
```

```python
merged.to_csv('test_textfooler_results_all.csv', index=False)
```
