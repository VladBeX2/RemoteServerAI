```python
import tensorflow as tf

# Setează GPU-ul vizibil înainte de orice altă operațiune TensorFlow
gpus = tf.config.list_physical_devices('GPU')
print(f"Available GPUs: {gpus}")

# Asigură-te că TensorFlow poate utiliza toate GPU-urile
for gpu in gpus:
    tf.config.experimental.set_memory_growth(gpu, True) 

# Creează strategia după ce dispozitivele au fost setate
strategy = tf.distribute.MirroredStrategy()

with strategy.scope():
    # Model simplu pentru test
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(512, activation='relu', input_shape=(1000,)),
        tf.keras.layers.Dense(10, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Datele de test
x = tf.random.uniform((12000, 1000))
y = tf.random.uniform((12000,), maxval=10, dtype=tf.int32)

# Antrenare
batch_size = 96
model.fit(x, y, epochs=3, batch_size=batch_size)
```
